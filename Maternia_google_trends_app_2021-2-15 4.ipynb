```python
# Standard imports

from pytrends.request import TrendReq
import pandas as pd
```

```python
pytrend = TrendReq(hl='cz', timeout=(100,100), tz=360)
```

```python
Keywords=['contact lenses','sunglasses','glasses'] # you can use any keywords such as "brýle", "sluneční brýle", "sunglassses", etc.
```

```python
COUNTRY=["GB"] #Use this link to get country code: https://en.wikipedia.org/wiki/List_of_ISO_3166_country_code
```

```python
Date_interval='today 3-m' # specify date range, you can also use "today 3-m" and "today 1-m" for last 3 and 1 months
```

```python
Category=0 # Use this link to select categories: https://github.com/pat310/google-trends-api/wiki/Google-Trends-Categories
```

```python
Search_type='' #default is 'web searches',others include 'images','news','youtube','froogle' (google shopping)
```

```python
Individual_keyword = list(zip(*[iter(Keywords)]*1))
Individual_keyword = [list(x) for x in Individual_keyword]
dicti = []

for Country in COUNTRY:
    for keyword in Individual_keyword:
        pytrend.build_payload(kw_list=keyword,
                              timeframe = Date_interval,
                              geo = Country,
                              cat=Category,
                              gprop=Search_type)
        dicti.append(pytrend.related_topics())

dataFrame = []
```

```python
for dict in dicti:
    for kw in dict:
        rising = dict[kw]['rising']

        for index, row in rising.iterrows():
            dataFrame.append([
                row.array[0],
                row.array[1],
                row.array[2],
                row.array[3],
                row.array[4],
                row.array[5],
                kw,
                'rising',
            ])

        top = dict[kw]['top']

        for index, row in top.iterrows():
            dataFrame.append([
                row.array[0],
                row.array[1],
                row.array[2],
                row.array[3],
                row.array[4],
                row.array[5],
                kw,
                'top',
            ])

dataFrame
```

```python
data = pd.DataFrame(dataFrame, columns=['value','change','url', 'url_2','category','topic', 'key_word','measure'])
```

```python
rising = data.loc[(data['measure'] == 'rising'), ['value','change','category','topic', 'key_word','measure']]
```
